```python
from __future__ import print_function, division
```

here is a mostly-straightforward solution to the<br>
two-by-two version of the grid.

```python
def do_twice(f):
    f()
    f()
```

```python
def do_four(f):
    do_twice(f)
    do_twice(f)
```

```python
def print_beam():
    print('+ - - - -', end=' ')
```

```python
def print_post():
    print('|        ', end=' ')
```

```python
def print_beams():
    do_twice(print_beam)
    print('+')
```

```python
def print_posts():
    do_twice(print_post)
    print('|')
```

```python
def print_row():
    print_beams()
    do_four(print_posts)
```

```python
def print_grid():
    do_twice(print_row)
    print_beams()
```

```python
print_grid()
    
```

here is a less-straightforward solution to the<br>
four-by-four grid

```python
def one_four_one(f, g, h):
    f()
    do_four(g)
    h()
```

```python
def print_plus():
    print('+', end=' ')
```

```python
def print_dash():
    print('-', end=' ')
```

```python
def print_bar():
    print('|', end=' ')
```

```python
def print_space():
    print(' ', end=' ')
```

```python
def print_end():
    print()
```

```python
def nothing():
    "do nothing"
```

```python
def print1beam():
    one_four_one(nothing, print_dash, print_plus)
```

```python
def print1post():
    one_four_one(nothing, print_space, print_bar)
```

```python
def print4beams():
    one_four_one(print_plus, print1beam, print_end)
```

```python
def print4posts():
    one_four_one(print_bar, print1post, print_end)
```

```python
def print_row():
    one_four_one(nothing, print4posts, print4beams)
```

```python
def print_grid():
    one_four_one(print4beams, print_row, nothing)
```

```python
print_grid()
```
