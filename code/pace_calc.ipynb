
<br>
If you run a 10 kilometer race in 43 minutes 30 seconds, what is your<br>
average time per mile? What is your average speed in miles per hour?<br>
(Hint: there are 1.61 kilometers in a mile).<br>


```python
minutes = 43.5
hours = minutes / 60
```

```python
km_per_mile = 1.61
km = 10
miles = km / km_per_mile 
```

```python
pace = minutes / miles
mph = miles / hours
```

```python
print('Pace in minutes per mile:', pace)
print('Average speed in mph:', mph)
```
