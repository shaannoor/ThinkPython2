```python
from __future__ import print_function, division
```

```python
import time
```

```python
def make_word_list1():
    """Reads lines from a file and builds a list using append."""
    t = []
    fin = open('words.txt')
    for line in fin:
        word = line.strip()
        t.append(word)
    return t
```

```python
def make_word_list2():
    """Reads lines from a file and builds a list using list +."""
    t = []
    fin = open('words.txt')
    for line in fin:
        word = line.strip()
        t = t + [word]
    return t
```

```python
start_time = time.time()
t = make_word_list1()
elapsed_time = time.time() - start_time
```

```python
print(len(t))
print(t[:10])
print(elapsed_time, 'seconds')
```

```python
start_time = time.time()
t = make_word_list2()
elapsed_time = time.time() - start_time
```

```python
print(len(t))
print(t[:10])
print(elapsed_time, 'seconds')
```
