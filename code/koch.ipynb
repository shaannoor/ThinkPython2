```python
from __future__ import print_function, division
```

```python
import turtle
```

```python
def koch(t, n):
    """Draws a koch curve with length n."""
    if n < 10:
        t.fd(n)
        return
    m = n/3
    koch(t, m)
    t.lt(60)
    koch(t, m)
    t.rt(120)
    koch(t, m)
    t.lt(60)
    koch(t, m)
```

```python
def snowflake(t, n):
    """Draws a snowflake (a triangle with a Koch curve for each side)."""
    for i in range(3):
        koch(t, n)
        t.rt(120)
```

```python
bob = turtle.Turtle()
```

```python
bob.pu()
bob.goto(-150, 90)
bob.pd()
snowflake(bob, 300)
```

```python
turtle.mainloop()
```
