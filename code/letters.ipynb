LEVEL 0 PRIMITIVES <br>
fd, bk, lt, rt, pu, pd

```python
def fd(t, length):
    t.fd(length)
```

```python
def bk(t, length):
    t.bk(length)
```

```python
def lt(t, angle=90):
    t.lt(angle)
```

```python
def rt(t, angle=90):
    t.rt(angle)
```

```python
def pd(t):
    t.pd()
```

```python
def pu(t):
    t.pu()
```

LEVEL 1 PRIMITIVES are simple combinations of Level 0 primitives.<br>
They have no pre- or post-conditions.

```python
def fdlt(t, n, angle=90):
    """forward and left"""
    fd(t, n)
    lt(t, angle)
```

```python
def fdbk(t, n):
    """forward and back, ending at the original position"""
    fd(t, n)
    bk(t, n)
```

```python
def skip(t, n):
    """lift the pen and move"""
    pu(t)
    fd(t, n)
    pd(t)
```

```python
def stump(t, n, angle=90):
    """Makes a vertical line and leave the turtle at the top, facing right"""
    lt(t)
    fd(t, n)
    rt(t, angle)
```

```python
def hollow(t, n):
    """move the turtle vertically and leave it at the top, facing right"""
    lt(t)
    skip(t, n)
    rt(t)
```

LEVEL 2 PRIMITIVES use primitives from Levels 0 and 1<br>
to draw posts (vertical elements) and beams (horizontal elements)<br>
Level 2 primitives ALWAYS return the turtle to the original<br>
location and direction.

```python
def post(t, n):
    """Makes a vertical line and return to the original position"""
    lt(t)
    fdbk(t, n)
    rt(t)
```

```python
def beam(t, n, height):
    """Makes a horizontal line at the given height and return."""
    hollow(t, n*height)
    fdbk(t, n)
    hollow(t, -n*height)
```

```python
def hangman(t, n, height):
    """Makes a vertical line to the given height and a horizontal line
    at the given height and then return.
    This is efficient to implement, and turns out to be useful, but
    it's not so semantically clean."""
    stump(t, n * height)
    fdbk(t, n)
    lt(t)
    bk(t, n*height)
    rt(t)
```

```python
def diagonal(t, x, y):
    """Makes a diagonal line to the given x, y offsets and return"""
    from math import atan2, sqrt, pi
    angle = atan2(y, x) * 180 / pi
    dist = sqrt(x**2 + y**2)
    lt(t, angle)
    fdbk(t, dist)
    rt(t, angle)
```

```python
def vshape(t, n, height):
    diagonal(t, -n/2, height*n)
    diagonal(t, n/2, height*n)
```

```python
def bump(t, n, height):
    """Makes a bump with radius n at height*n 
    """
    stump(t, n*height)
    arc(t, n/2.0, 180)
    lt(t)
    fdlt(t, n*height+n)
```


<br>
The letter-drawing functions all have the precondition<br>
that the turtle is in the lower-left corner of the letter,<br>
and postcondition that the turtle is in the lower-right<br>
corner, facing in the direction it started in.<br>
They all take a turtle as the first argument and a size (n)<br>
as the second.  Most letters are (n) units wide and (2n) units<br>
high.<br>


```python
def draw_a(t, n):
    diagonal(t, n/2, 2*n)
    beam(t, n, 1)
    skip(t, n)
    diagonal(t, -n/2, 2*n)
```

```python
def draw_b(t, n):
    bump(t, n, 1)
    bump(t, n, 0)
    skip(t, n/2)
```

```python
def draw_c(t, n):
    hangman(t, n, 2)
    fd(t, n)
```

```python
def draw_d(t, n):
    bump(t, 2*n, 0)
    skip(t, n)
```

```python
def draw_ef(t, n):
    hangman(t, n, 2)
    hangman(t, n, 1)
```

```python
def draw_e(t, n):
    draw_ef(t, n)
    fd(t, n)
```

```python
def draw_f(t, n):
    draw_ef(t, n)
    skip(t, n)
```

```python
def draw_g(t, n):
    hangman(t, n, 2)
    fd(t, n/2)
    beam(t, n/2, 2)
    fd(t, n/2)
    post(t, n)
```

```python
def draw_h(t, n):
    post(t, 2*n)
    hangman(t, n, 1)
    skip(t, n)
    post(t, 2*n)
```

```python
def draw_i(t, n):
    beam(t, n, 2)
    fd(t, n/2)
    post(t, 2*n)
    fd(t, n/2)
```

```python
def draw_j(t, n):
    beam(t, n, 2)
    arc(t, n/2, 90)
    fd(t, 3*n/2)
    skip(t, -2*n)
    rt(t)
    skip(t, n/2)
```

```python
def draw_k(t, n):
    post(t, 2*n)
    stump(t, n, 180)
    vshape(t, 2*n, 0.5)
    fdlt(t, n)
    skip(t, n)
```

```python
def draw_l(t, n):
    post(t, 2*n)
    fd(t, n)
```

```python
def draw_n(t, n):
    post(t, 2*n)
    skip(t, n)
    diagonal(t, -n, 2*n)
    post(t, 2*n)
```

```python
def draw_m(t, n):
    post(t, 2*n)
    draw_v(t, n)
    post(t, 2*n)
```

```python
def draw_o(t, n):
    skip(t, n)
    circle(t, n)
    skip(t, n)
```

```python
def draw_p(t, n):
    bump(t, n, 1)
    skip(t, n/2)
```

```python
def draw_q(t, n):
    draw_o(t, n)
    diagonal(t, -n/2, n)
```

```python
def draw_r(t, n):
    draw_p(t, n)
    diagonal(t, -n/2, n)
```

```python
def draw_s(t, n):
    fd(t, n/2)
    arc(t, n/2, 180)
    arc(t, n/2, -180)
    fdlt(t, n/2, -90)
    skip(t, 2*n)
    lt(t)
```

```python
def draw_t(t, n):
    beam(t, n, 2)
    skip(t, n/2)
    post(t, 2*n)
    skip(t, n/2)
```

```python
def draw_u(t, n):
    post(t, 2*n)
    fd(t, n)
    post(t, 2*n)
```

```python
def draw_v(t, n):
    skip(t, n/2)
    vshape(t, n, 2)
    skip(t, n/2)
```

```python
def draw_w(t, n):
    draw_v(t, n)
    draw_v(t, n)
```

```python
def draw_x(t, n):
    diagonal(t, n, 2*n)
    skip(t, n)
    diagonal(t, -n, 2*n)
```

```python
def draw_v(t, n):
    skip(t, n/2)
    diagonal(t, -n/2, 2*n)
    diagonal(t, n/2, 2*n)
    skip(t, n/2)
```

```python
def draw_y(t, n):
    skip(t, n/2)
    stump(t, n)
    vshape(t, n, 1)
    rt(t)
    fdlt(t, n)
    skip(t, n/2)
```

```python
def draw_z(t, n):
    beam(t, n, 2)
    diagonal(t, n, 2*n)
    fd(t, n)
```

```python
def draw_(t, n):
    # draw a space
    skip(t, n)
```
